# ***Gaussian Naive Bayes***

Gaussian Naive Bayes (GNB) is a probabilistic classification algorithm based on Bayes’ Theorem with a strong assumption that the features follow a normal (Gaussian) distribution and are conditionally independent given the class label.

It is one of the simplest and fastest classification algorithms used especially in text classification, medical data analysis, and real-time prediction systems, yet it can perform surprisingly well even with strong feature correlation.
image.png

```python
#Let's split the data
from sklearn.model_selection import train_test_split

X = df.drop(['target_name', 'target'], axis=1) # Drop both 'target_name' and 'target' from features
y = df['target_name']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#Define Gaussian PDF Function
def gaussian_pdf(x, mean, var):
    eps = 1e-9  # to avoid division by zero
    coefficient = 1.0 / np.sqrt(2.0 * np.pi * var + eps)
    exponent = np.exp(- (x - mean) ** 2 / (2 * var + eps))
    return coefficient * exponent

```

```python
#Train the GNB Model
class GaussianNaiveBayesScratch:
    def fit(self, X, y):
        self.classes = np.unique(y)
        self.params = {}
        self.priors = {}

        for cls in self.classes:
            X_c = X[y == cls]
            self.params[cls] = {
                'mean': X_c.mean(axis=0),
                'var': X_c.var(axis=0)
            }
            self.priors[cls] = X_c.shape[0] / X.shape[0]

    def predict(self, X):
        # Check if X is a pandas DataFrame and use .values if it is
        if isinstance(X, pd.DataFrame):
            y_pred = [self._predict_sample(x) for x in X.values] # Iterate over the rows (samples)
        else:
            y_pred = [self._predict_sample(x) for x in X] # Assume X is already a numpy array and iterate directly
        return np.array(y_pred)


    def _predict_sample(self, x):
        posteriors = []

        for cls in self.classes:
            prior = np.log(self.priors[cls])
            class_conditional = np.sum(np.log(gaussian_pdf(x,
                                                           self.params[cls]['mean'],
                                                           self.params[cls]['var'])))
            posterior = prior + class_conditional
            posteriors.append(posterior)

        return self.classes[np.argmax(posteriors)]
```

```python
#Train and Evaluate the model
# Instantiate and train

gnb_scratch = GaussianNaiveBayesScratch()
gnb_scratch.fit(X_train, y_train)

# Predict
y_pred = gnb_scratch.predict(X_test)

# Evaluate
accuracy = np.mean(y_pred == y_test)
print(f"Accuracy1: {accuracy:.2f}")
```

```python
#Classification report and confusion Matrix
from sklearn.metrics import classification_report, confusion_matrix
cm1= confusion_matrix(y_test, y_pred)
print(cm1)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=le.classes_))
```
